```python
sample_text = """
How do I reset my password?
To reset your password, click on the 'Forgot Password' link on the login page.
You will receive a password reset email within a few minutes.
If you do not receive the email, check your spam folder.

Refund Policy
Refunds are processed within 5–7 business days after approval.
Refunds are not available for discounted or promotional plans.
Once a refund is processed, it may take additional time for the bank to reflect the amount.

Account Deletion
You can request account deletion by contacting support.
Once deleted, your data cannot be recovered.
"""
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter


splitter = RecursiveCharacterTextSplitter(
    chunk_size=300,
    chunk_overlap=50,
    separators=["\n\n", "\n", ".", " ", ""]
)

chunks = splitter.split_text(sample_text)
```

```python
for i, chunk in enumerate(chunks):
    print(f"\n--- Chunk {i} (length={len(chunk)}) ---\n")
    print(chunk)
```

```python
documents = []

for i, chunk in enumerate(chunks):
    documents.append({
        "content": chunk,
        "metadata": {
            "tenant_id": "test_company",
            "source": "faq",
            "chunk_index": i
        }
    })

documents
```
